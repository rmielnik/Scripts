# Psycho data generator

Provide phrase length and a number:

```python
phrase = [3, 2, 3] # phrase OooOoOoo
accentDistances = [3, 7, 9] # generate three sequences with accent every 3, 7 and 9 notes respectively 
```

```python
import math

for accentDistance in accentDistances:
    fullPhraseLength = sum(phrase) 
    if (math.gcd(fullPhraseLength, accentDistance) != 1):
        raise Exception("full phrase length and accent distance must be coprime integers")

    fullAccent = ('▼' + '.' * (accentDistance - 1))
    fullPhrase = ''.join(map(lambda x: 'O' + 'o' * (x - 1), phrase))

    # generate whole sequence
    repetitionIndicator = ''.join([str(x).ljust(fullPhraseLength) for x in range(1, accentDistance + 1)])
    phraseSequence = fullPhrase * accentDistance
    accentSequence = fullAccent * fullPhraseLength
    fullSequenceLength = len(phraseSequence)

    #Uncomment to print whole squence
    #print(repetitionIndicator)
    #print(phraseSequence)
    #print(accentSequence)

    # compact 
    head = 0
    tail = 0 

    compactedRepetitionIndicator = ''
    compactedPhraseSequence = ''
    compactedAccentSequence = ''

    while head < fullSequenceLength:
        if all(accentSequence[x] == '.' for x in range(head, head + fullPhraseLength)):
            counter = 1
            tail = head + fullPhraseLength

            while tail + fullPhraseLength <= fullSequenceLength and all(accentSequence[x] == '.' for x in range (tail, tail + fullPhraseLength)):
                tail += fullPhraseLength
                counter += 1

            counterString = ' [' + str(counter) + '] '
            compactedRepetitionIndicator += ' ' * len(counterString)
            compactedPhraseSequence += counterString
            compactedAccentSequence += ' ' * len(counterString)
            head = tail
        else:
            compactedRepetitionIndicator += repetitionIndicator[head : head + fullPhraseLength]
            compactedPhraseSequence += phraseSequence[head : head + fullPhraseLength]
            compactedAccentSequence += accentSequence[head : head + fullPhraseLength]
            head += fullPhraseLength
            tail = head

    print(compactedRepetitionIndicator)
    print(compactedPhraseSequence)
    print(compactedAccentSequence)
```
